```python
import cv2
import glob
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_size_inches(10,10)

impath = glob.glob("polygon-yolov5/runs/detect/exp65/*.png")
for path in impath[5:15]:
    # path = impath[2]
    label = path.replace(os.path.basename(path),"labels/"+os.path.basename(path).replace('.png','.txt'))

    if os.path.isfile(label):
        with open(label,'r') as f:
            anos = f.readlines()
    else:
        continue

    img = cv2.imread(path)
    plt.imshow(img)
    plt.show()

    for ano in anos:
    #     ano = anos[0]
        ano = ano.replace('\n','')
        ano = ano.split(' ')
        ano = np.array(ano, dtype=np.float32)

        ## edge에 가까운 것 제거
        if any(ano[1:] < 0.01):
            continue
        if any(ano[1:] > 0.99):
            continue

        ano[1::2] = ano[1::2] * img.shape[1]
        ano[::2] = ano[::2] * img.shape[0]
        ano = ano.astype(np.int32)



        point1, point2, point3 = ano[1:3], ano[3:5], ano[5:7]
        cv2.line(img, point1,point1,(255,0,0),25)
        cv2.line(img, point2,point2,(255,127,0),25)
        cv2.line(img, point3,point3,(255,255,0),25)

        new_w = int(np.linalg.norm(point1-point2))
        new_h = int(np.linalg.norm(point2-point3))

        pts1 = np.float32([point1,point2,point3])
        pts2 = np.float32([[0,0], [new_w, 0], [new_w, new_h]])
        mat = cv2.getAffineTransform(pts1,pts2)
        dst = cv2.warpAffine(img, mat, None)
        dst = dst[:new_h,:new_w]

        plt.imshow(dst)
        plt.show()
```
